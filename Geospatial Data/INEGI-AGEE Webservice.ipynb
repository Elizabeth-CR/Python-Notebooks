```python
from os import sched_getscheduler
import json
import urllib.request
import geoplot
import geoplot.crs as gcrs
import folium
import webbrowser
import findspark
```

```python
# Initialize findspark
findspark.init()

# Create a PySpark session
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
sc = SparkContext.getOrCreate()
spark = SparkSession(sc)
spark

with urllib.request.urlopen('https://gaia.inegi.org.mx/wscatgeo/geo/mgee/') as f:
    data = json.loads(f.read().decode())
    df = spark.createDataFrame(data['features'])
df.show()
```

```python
df = df.dropna(subset=['geometry'])
display(df.head())
```

```python
states = ("https://gaia.inegi.org.mx/wscatgeo/geo/mgee/")
m = folium.Map(location=(30, 10), zoom_start=3, tiles="OpenStreetMap")
folium.GeoJson(states).add_to(m)
m.save("footprint.html")
display(m)
```
